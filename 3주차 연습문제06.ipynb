```python
import numpy as np
import matplotlib.pyplot as plt

def gaussian(x, mu, sigma):
    return np.exp(-(x - mu)**2 / (2 * sigma**2))

x = np.linspace(-5, 5, 100)
y = gaussian(x, 0, 1)

plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('y')
plt.title('1D Gaussian Function')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

def gaussian2d(x, y, mu, sigma):
    return np.exp(-(x - mu[0])**2 / (2 * sigma[0]**2)) * np.exp(-(y - mu[1])**2 / (2 * sigma[1]**2))

x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x, y)
Z = gaussian2d(X, Y, [0, 0], [1, 1])

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z, cmap='viridis')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
ax.set_title('2D Gaussian Function')
plt.show()
```
